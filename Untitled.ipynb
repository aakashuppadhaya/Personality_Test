```python
from sklearn.preprocessing import LabelEncoder
lablel_encoder = LabelEncoder()
X['Gender'] = lablel_encoder.fit_transform(X['Gender'])
X['Married'] = lablel_encoder.fit_transform(X['Married'])
X['Self_Employed'] = lablel_encoder.fit_transform(X['Self_Employed'])
X['Property_Area'] = lablel_encoder.fit_transform(X['Property_Area'])
```

```python
X['Education'] = lablel_encoder.fit_transform(X['Education'])
```

```python
X['Dependents'] = lablel_encoder.fit_transform(X['Dependents'])
```

```python
X.info()
```

```python
Y= lablel_encoder.fit_transform(Y)
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train_stratify, X_test_stratify, y_train_stratify, y_test_stratify = train_test_split(X,Y, test_size=0.2, random_state=42,stratify=Y)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, random_state=5, n_jobs=-1)
clf.fit(X_train_stratify, y_train_stratify)
feat_importances = pd.Series(clf.feature_importances_, index=X_train_stratify.columns)
feat_importances.nlargest(14).plot(kind='barh')
plt.show()
```

```python
from sklearn.metrics import f1_score, confusion_matrix
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import roc_auc_score
y_pred = clf.predict(X_test_stratify)

accuracy = clf.score(X_test_stratify,y_test_stratify)

print(accuracy)

precision = precision_score(y_test_stratify,y_pred)

print(precision)

recall = recall_score(y_test_stratify,y_pred)

print(recall)

f1 = f1_score(y_test_stratify,y_pred)

print(f1)

confusion_mat = confusion_matrix(y_test_stratify,y_pred)

print(confusion_mat)
```

```python
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(y_test_stratify, clf.predict(X_test_stratify), pos_label=1)
roc_auc=metrics.auc(fpr, tpr)
```

```python
# ROC curve with visualization
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(y_test_stratify, y_pred, pos_label=1)
roc_auc=metrics.auc(fpr, tpr)
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
